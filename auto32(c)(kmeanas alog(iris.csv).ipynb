```python
#kmeans algorithm
import numpy as np
import pandas  as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('iris.csv', names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'])
df.head()
```

```python
X = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = df['species']
```

```python
df.head()
```

```python
df = df.drop('species',axis=1)
df.head()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
inertia = []
for i in range(1,5):
    kmean = KMeans(n_clusters=i,random_state=0,n_init="auto")
    kmean.fit(df)
    inertia.append(kmean.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,5),inertia)
plt.show()
```

```python
kmean = KMeans(n_clusters=3,random_state=0,n_init="auto")
```
